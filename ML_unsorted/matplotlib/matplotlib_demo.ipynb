## What is matplotlib?

[Matplotlib](https://matplotlib.org) це бібліотека візуалізації для Python.

Наприклад, якщо ви хочете відобразити щось на діаграмі чи графіку, matplotlib може допомогти вам зробити це програмно.

### Anatomy of a Matplotlib Figure

Matplotlib пропонує практично необмежені можливості для створення графіків.

Однак давайте розберемо деякі основні терміни.

* **Figure** - Основне полотно всіх графіків matplotlib. Загальна річ, яку ви малюєте, — це фігура, яку часто скорочують до `fig`.
* **Axes** - Одна фігура може мати одну *або* кілька осей, наприклад, фігура з кількома підсхемами може мати 4 осі (2 рядки та 2 стовпці). Часто скорочується до `ax`.
* **Axis** - Певний розмір осей, наприклад, вісь x або вісь y.

### A quick Matplotlib Workflow

Наступний робочий процес є стандартною практикою під час створення діаграми matplotlib:

0. **Import matplotlib** - Як приклад, `import matplotlib.pyplot as plt`.
1. **Prepare data** - Це може бути з наявного набору даних (аналіз даних) або з результатів моделі машинного навчання.
2. **Setup the plot** - Іншими словами, створіть фігуру та різні осі.
3. **Plot data to the Axes** - Надішліть відповідні дані до цільових осей.
4. **Customize the plot** - Додайте заголовок, прикрасьте кольори, позначте кожну вісь.
5. **Save (optional) and show** - Подивіться, як виглядає ваш графік, і за потреби збережіть його у файлі.

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
# %matplotlib inline # Not necessary in newer versions of Jupyter (e.g. 2022 onwards)
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot (Figure and Axes)
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("simple-plot.png")
```

### Creating Figures with multiple Axes with Subplots

Subplots дозволяють створювати кілька осей на одній фігурі.

Subplots корисні, оскільки ви починаєте з однієї ділянки на фігуру, але за потреби збільшуєте її до більшої кількості.

Наприклад, давайте створимо Subplots, яка показує багато з наведених вище наборів даних на одному малюнку.

Ми можемо зробити це, створивши кілька осей за допомогою [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots) і встановлення параметрів `nrows` (кількість рядків) і `ncols` (кількість стовпців), щоб відобразити, скільки осей ми хочемо.

Параметри `nrows` і `ncols` є мультиплікативними, тобто `plt.subplots(nrows=2, ncols=2)` створить 2*2=4` загальних осей.

```python
# Option 1: Create 4 subplots with each Axes having its own variable name
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2, 
    ncols=2, 
    figsize=(10, 5)
)

# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2: Create 4 subplots with a single ax variable
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Index the ax variable to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Line plot from a pandas DataFrame

```python
# Start with some dummy data
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2024', periods=1000))

# Note: ts = short for time series (data over time)
ts
```

Great! We've got some random values across time.

Now let's add up the data cumulatively overtime with [`DataFrame.cumsum()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cumsum.html) (`cumsum` is short for cumulative sum or continaully adding one thing to the next and so on). 

```python
# Add up the values cumulatively
ts.cumsum()
```

We can now visualize the values by calling the `plot()` method on the DataFrame and specifying the kind of plot we'd like with the `kind` parameter.

In our case, the kind we'd like is a line plot, hence `kind="line"` (this is the default for the `plot()` method).

```python
# Plot the values over time with a line plot (note: both of these will return the same thing)
# ts.cumsum().plot() # kind="line" is set by default
ts.cumsum().plot(kind="line");
```

### Plotting more advanced plots from a pandas DataFrame

З pandas DataFrame можна отримати набагато складніші та детальніші графіки.

Давайте потренуємося використовувати `heart_disease` DataFrame.

І як приклад, давайте проаналізуємо людей старше 50 років.

Для цього давайте почнемо зі створення графіка безпосередньо з pandas, а потім за допомогою `plt.subplots()`.

```python
# Perform data analysis on patients over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

Тепер давайте створимо діаграму розсіювання безпосередньо з pandas DataFrame.

Це досить легко зробити, але дещо обмежено з точки зору налаштування.

Давайте візуалізуємо пацієнтів із рівнем холестерину понад 50.

Ми можемо візуалізувати, які пацієнти мають або не мають захворювання серця, розфарбувавши зразки відповідно до цільового стовпця `target` (наприклад, `0` = відсутність захворювання серця, `1` = захворювання серця).

```python
# Create a scatter plot directly from the pandas DataFrame
over_50.plot(
    kind="scatter",
    x="age", 
    y="chol", 
    c="target", # colour the dots by target value
    figsize=(10, 6)
);
```

Ми можемо відтворити той самий графік за допомогою `plt.subplots()`, а потім передати змінну Axes (`ax`) методу pandas `plot()`.

```python
# Create a Figure and Axes instance
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data from the DataFrame to the ax object
over_50.plot(kind="scatter", 
             x="age", 
             y="chol", 
             c="target", 
             ax=ax); # set the target Axes

# Customize the x-axis limits (to be within our target age ranges)
ax.set_xlim([45, 100]);
```

Тепер замість того, щоб побудувати графік безпосередньо з pandas DataFrame, ми можемо зробити трохи більш комплексний графік, наносячи дані безпосередньо на цільовий екземпляр Axes.

```python
# Create Figure and Axes instance
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data directly to the Axes intance
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"]) # Colour the data with the "target" column

# Customize the plot parameters 
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Setup the legend
ax.legend(*scatter.legend_elements(), 
          title="Target");
```

Що, якби ми хотіли, щоб горизонтальна лінія перетинала середнє значення `heart_disease["chol"]`?

Ми робимо це за допомогою методу `Axes.axhline()`.

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legned
ax.legend(*scatter.legend_elements(), 
          title="Target")

# Add a meanline
ax.axhline(over_50["chol"].mean(),
           linestyle="--"); # style the line to make it look nice
```

### Plotting multiple plots on the same figure

Іноді вам потрібно візуалізувати кілька функцій набору даних або результатів моделі на одному малюнку.

Ви можете досягти цього, додавши дані до кількох осей на одному малюнку.

Метод `plt.subplots()` допомагає вам створювати фігури з бажаною кількістю осей у бажаній фігурі.

Використовуючи параметри `nrows` (кількість рядків) і `ncols` (кількість стовпців), ви можете контролювати кількість осей на малюнку.

Наприклад:

- nrows=2, ncols=1 = 2x1 = фігура з 2 осями
- nrows=5, ncols=5 = 5x5 = фігура з 25 осями

Давайте створимо графік з 2 осями.

Однією з перших осей (осі 0) ми будемо порівнювати захворювання серця з рівнем холестерину (`chol`).

На другій осі (вісь 1) ми відобразимо хворобу серця порівняно з максимальним рівнем частоти серцевих скорочень (`thalach`).

```python
# Setup plot (2 rows, 1 column)
fig, (ax0, ax1) = plt.subplots(nrows=2, # 2 rows
                               ncols=1, # 1 column 
                               sharex=True, # both plots should use the same x-axis 
                               figsize=(10, 8))

# ---------- Axis 0: Heart Disease and Cholesterol Levels ----------

# Add data for ax0
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='#9c49bf', 
            linestyle='--', 
            label="Average")

# ---------- Axis 1: Heart Disease and Max Heart Rate Levels ----------

# Add data for ax1
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='g', 
            linestyle='--', 
            label="Average")

# Title the figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, 
             fontweight='bold');
```

Якщо у вас є об’єкт Figure або Axes matplotlib, ви можете налаштувати багато атрибутів за допомогою методу `Axes.set()`.

Наприклад, ви можете змінити:

- `xlabel` - мітки на осі x.
- `ylim` - межі осі у.
- `xticks` - Стиль x-ticks.

Спочатку ми створимо деякі випадкові дані, а потім помістимо їх у DataFrame.

Потім ми створимо графік із цього DataFrame і побачимо, як його налаштувати.

```python
# Create random data
x = np.random.randn(10, 4)
x
```

```python
# Turn data into DataFrame with simple column names
df = pd.DataFrame(
    x, 
    columns=['a', 'b', 'c', 'd']
)

df
```

Тепер давайте нанесемо дані з DataFrame на гістограму.

Цього разу ми збережемо графік у змінній під назвою `ax` (скорочення від Axes).

```python
# Create a bar plot
ax = df.plot(kind="bar")

# Check the type of the ax variable
type(ax)
```

Ми бачимо, що тип нашої змінної `ax` — `AxesSubplot`, що дозволяє нам використовувати всі методи, доступні в matplotlib для `Axes`.

Давайте встановимо кілька атрибутів графіка за допомогою методу `set()`.

Зокрема, ми змінимо `title`, `xlabel` і `ylabel`.

Зауважте, ми можемо змінити також місце розташування легенди, якщо хочемо, за допомогою параметра `loc` методу `legend()`.

`loc` можна встановити як рядок, щоб відобразити, де має бути легенда.

За замовчуванням встановлено значення `loc="best"`, що означає, що matplotlib намагатиметься визначити найкраще позиціонування для нього.

Давайте спробуємо змінити його на `"loc="upper left"`.

```python
# Recreate the ax object
ax = df.plot(kind="bar")

# Set various attributes
ax.set(title="Random Number Bar Graph from DataFrame", 
       xlabel="Row number", 
       ylabel="Random number")

# Change the legend position
ax.legend(loc="lower left");
```

Це краще підходить?

### Customizing the colours of plots with colormaps

Колір — одна з найважливіших ознак графіку.

Це може допомогти розділити різні види інформації.

Matplotlib надає багато різних варіантів кольорів через `matplotlib.colormaps`.

Давайте подивимося, як ми можемо змінити кольори діаграми matplotlib за допомогою параметра `cmap` (`cmap` — скорочення від `colormaps`).

Ми почнемо зі створення діаграми розсіювання зі значенням `cmap` за замовчуванням (`cmap="viridis"`).

```python
# Setup the Figure and Axes
fig, ax = plt.subplots(figsize=(10, 6))


# Create a scatter plot with no cmap change (use default colormap)
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"],
                     cmap="viridis") # default cmap value

# Add attributes to the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

ax.axhline(y=over_50["chol"].mean(), 
           c='b', 
           linestyle='--', 
           label="Average");

ax.legend(*scatter.legend_elements(), 
          title="Target");
```

Що, якби ми хотіли змінити кольори?

Спробуємо `cmap="winter"`.

Ми також можемо змінити колір горизонтальної лінії за допомогою параметра `color` і встановити для нього рядок бажаного кольору (наприклад, `color="r"` для червоного).

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Setup scatter plot with different cmap
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], 
                     cmap="cool") # Change cmap value 

# Add attributes to the plot with different color line
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

ax.axhline(y=over_50["chol"].mean(), 
           color="r", # Change color of line to "r" (for red)
           linestyle='--', 
           label="Average");

ax.legend(*scatter.legend_elements(), 
          title="Target");
```
